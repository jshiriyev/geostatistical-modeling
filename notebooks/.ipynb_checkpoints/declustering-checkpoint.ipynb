```python
class Mesh():

  def __init__(self,xmin,xmax,ymin,ymax):
    self.xmin = xmin
    self.xmax = xmax
    self.ymin = ymin
    self.ymax = ymax

  def grid(self,Nx,Ny):
    Xnodes = np.linspace(self.xmin,self.xmax,Nx+1)
    Ynodes = np.linspace(self.ymin,self.ymax,Ny+1)
    return Xnodes,Ynodes

  def size(self,Nx,Ny):
    dx = (self.xmax-self.xmin)/Nx
    dy = (self.ymax-self.ymin)/Ny
    return dx*dy

  def dcmean(self,Nx,Ny,X,Y,Z,returnFlag=False):
    """if returnFlag is True, the number of empty cells are returned too."""
    declustered_mean = np.zeros((Nx,Ny))

    Xnodes,Ynodes = self.grid(Nx,Ny)

    number_of_empty_cells = 0

    for i in range(Nx):

      for j in range(Ny):

        Xcond = np.logical_and(X>Xnodes[i],Xnodes[i+1]>X)
        Ycond = np.logical_and(Y>Ynodes[j],Ynodes[j+1]>Y)
        Scond = np.logical_and(Xcond,Ycond)

        if Z[Scond].size==0:
          declustered_mean[j,i] = np.nan
          number_of_empty_cells += 1
        else:
          declustered_mean[j,i] = Z[Scond].mean()

    if returnFlag:
      return declustered_mean,number_of_empty_cells

    return declustered_mean
```

```python
mesh = Mesh(xmin,xmax,ymin,ymax)

gcount,gsizes,gmeans = [],[],[]

for count in range(1,21):
  gcount.append(count)
  gsizes.append(np.sqrt(mesh.size(count,count)))
  gmeans.append(np.nanmean(mesh.dcmean(count,count,X,Y,PHI)))
```

```python
plt.scatter(gcount,gmeans)
plt.hlines(
    y = PHI.mean(),
    xmin = min(gcount),
    xmax = max(gcount),
    color = 'k',
    linestyle = '--',
    linewidth = 0.7,
    label = 'Actual Mean'
    )

plt.ylabel("Declustered Mean")

plt.legend()

plt.xticks(range(1,21))

plt.grid(axis="x")

plt.show()
```

```python
Nx,Ny = 9,9

Xnodes,Ynodes = mesh.grid(Nx,Ny)

plt.scatter(X,Y,s=10,c=PHI)
plt.colorbar()

for i in range(Nx+1):
  plt.vlines(Xnodes[i],ymin,ymax,color='k',linewidth=0.5)

for j in range(Ny+1):
  plt.hlines(Ynodes[j],xmin,xmax,color='k',linewidth=0.5)
```
